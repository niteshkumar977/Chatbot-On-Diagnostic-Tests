```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Convert to DataFrame
df = pd.read_csv('age_gender_symptoms_habits_real_tests.csv')
df.head()
```

```python
df_sym_hab = pd.read_csv('symptom_habit_master.csv')
df_sym_hab.head()
```

```python


# Function to combine features into a single string
def combine_features(row):
    return f"{row['age']} {row['gender']} {row['symptoms']} {row['habits']}"

# Apply combine_features function to the dataframe
df['Combined'] = df.apply(combine_features, axis=1)

# Function to recommend tests based on user input
def recommend_tests(age, gender, symptoms, habits):
    user_input = f"{age} {gender} {symptoms} {habits}"
    corpus = df['Combined'].tolist() + [user_input]
    #print (corpus)
    # Calculate TF-IDF vectors
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(corpus)
    #print (tfidf_matrix)
    #print("####")
    #print(tfidf_matrix[-1])
    #print("####")
    #print(tfidf_matrix[:-1])
    # Calculate cosine similarity
    cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])
    #print (cosine_similarities)
    # Get the most similar record
    idx = cosine_similarities.argsort()[0][-1]  #---COMMENTED FOR TESTING
    #top_indices = cosine_similarities[0].argsort()[-3:][::-1]  #----NEW INTRODUCTION
    
    # Get the tests of the top 3 most similar records 
    #top_tests = df.iloc[top_indices]['tests'].tolist()
    
    
    
    # Convert the list of top tests to a comma-separated string 
    #return ', '.join(top_tests)     #----NEW INTRODUCTION
    
    # Recommend the test
    return df.iloc[idx]['tests']     #---COMMENTED FOR TESTING

# Example usage
#age = 48
#gender = 'male'
#symptoms = 'Dizz'
#habits = 'hydration','poor diet'
#recommended_test = recommend_tests(age, gender, symptoms, habits)
#print(f"Recommended Test: {recommended_test}")
```

```python
# Function to print options based on input symptoms

def print_symptom_names(symptoms):
    symp_names = []
    for symptom in symptoms:
        options = df_sym_hab[df_sym_hab['symptom_no'] == symptom]['symptoms']
        for option in options:
            symp_names.append(option)
    #print(','.join(f"'{str(item)}'" for item in list(set(symp_names))))
    #return ','.join(f"'{str(item)}'" for item in list(set(symp_names)))
    print (tuple(set(symp_names)))
    return tuple(set(symp_names))
```

```python
# Function to print options based on input symptoms


def print_habits(symptoms):
    symp_name = []
    habit_no = []
    habit_name = []
    count = 0
    for symptom in symptoms:
        options = df_sym_hab[df_sym_hab['symptom_no'] == symptom]['habits'].tolist()
        #print(f"For symptom '{symptom}', the related habits are:") -----DEACTIVATED FOR TESTING
        for option in options:
            symp_name.append(symptom)
            
            count = count+1
            habit_no.append(count)
            
            habit_name.append(option)
            
           # print(f" - {option}") --- DEACTIVATED FOR TESTING
        #print("\n")
    
    habit_df = pd.DataFrame({'symptoms' : symp_name,'habit_no' : habit_no,'habit_name' : habit_name})
    
    df_sym_unique = df_sym_hab[['symptom_no','symptoms']]
    df_sym_unique = df_sym_unique.drop_duplicates()
    
    
    habit_df = pd.merge(habit_df,df_sym_unique,left_on = 'symptoms',right_on = 'symptom_no',how = 'inner')
    
    habit_df = habit_df.drop(columns = ['symptoms_x','symptom_no'])
    
    habit_df = habit_df.rename(columns ={'symptoms_y' : 'symptom_name'})
    habit_df = habit_df[['habit_no','symptom_name','habit_name']]
    
    #habit_df(index = False))
    #print(habit_df.to_string(index = False))
    print(habit_df)
    return habit_df
    
```

```python
# Function to print options based on input sympto

def print_habit_names(habits,habit_dataframe):
    hab_names = []
    for habit in habits:
        options = habit_dataframe[habit_dataframe['habit_no'] == habit]['habit_name']
        for option in options:
            hab_names.append(option)
    #print(','.join(f"'{str(item)}'" for item in list(hab_names)))
    #eval(f"({string_data})")
    print (tuple(hab_names))
    return tuple(hab_names)
    #return ','.join(f"'{str(item)}'" for item in list(hab_names))
```
